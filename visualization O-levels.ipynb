```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_excel("results_gpt for O Levels/Results analysis_Olevels all results_essay ids.xlsx")


df_2=df.dropna(subset=['human2_Score'])
print(len(df_2))
df.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Sample scores assigned by reader1, human2, and GPT
human1_scores = df_2['human1_Score'].astype(int).tolist()
# df['human2_Score'] = df['human2_Score'].replace('NA', '0')
human2_scores = df_2['human2_Score'].astype(int).tolist()
gpt_scores = df_2['GPT_Score'].astype(int).tolist()

# Sample data - essay ID and scores
essay_id = df_2['Essay ID']

# Set the width of each bar
bar_width = 0.2

# Set the positions of the bars on the x-axis
bar_positions = np.arange(len(essay_id))
plt.figure(figsize=(6,5))
# Plotting the scores for human1, human2, and GPT
plt.bar(bar_positions, human1_scores, width=bar_width, align='center', label='Human1')
plt.bar(bar_positions + bar_width, human2_scores, width=bar_width, align='center', label='Human2')
plt.bar(bar_positions + 2*bar_width, gpt_scores, width=bar_width, align='center', label='GPT')

# Adding labels and title to the graph
plt.xlabel('Essay ID')
plt.ylabel('Scores')
plt.title('GPT versus Human Assessers')

# Adding x-axis tick labels
plt.xticks(bar_positions + bar_width, essay_id)
plt.yticks(range(0, 31, 5))
# Adding a legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.18), ncol=3)

plt.tight_layout()
plt.savefig("O-Levels Score human2")

# Display the graph
plt.show()
```
